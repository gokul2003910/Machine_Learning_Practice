Linear Regression, Ridge and lasso

```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = fetch_california_housing()
```

```python
type(df)
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
dataset['price']=df.target
dataset.head()
```

```python
#  Dividing the Dataset into independent and dependent features
X = dataset.iloc[:, :-1]  # Independent features
y = dataset.iloc[:, -1]   # Dependent feature (price)
```

<!-- Linear Regression -->

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg= LinearRegression()
mse= cross_val_score(lin_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
mse
```

```python
mean_mse = np.mean(mse)
mean_mse
```

```python
lin_reg.fit(X, y)
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

params = {'alpha': [1e-15, 1e-10, 1e-5, 0.1, 1, 10, 20, 30, 40, 50, 80, 100]}
ridge_reg = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=5)
ridge_reg.fit(X_train, y_train)
```

```python
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()

params = {'alpha': [1e-15, 1e-10, 1e-5, 0.1, 1, 10, 20, 30, 40, 50,80, 100]}
lasso_reg = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=20)
lasso_reg.fit(X_train, y_train)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

```python
y_pred = lin_reg.predict(X_test)
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(r2)
```
